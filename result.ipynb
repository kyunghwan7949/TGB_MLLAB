```python
import matplotlib.pyplot as plt

epoch = [0, 1, 2, 3, 4]
valid_mrr = [0, 0.1104, 0.1538, 0.3795, 0.4357]
valid_tmp_mrr = [0, 0.008, 0.01, 0.0015, 0.002]

# Plotting valid_mrr
plt.plot(epoch, valid_mrr, marker='o', linestyle='-', label='valid_mrr')

# Plotting valid_tmp_mrr
plt.plot(epoch, valid_tmp_mrr, marker='x', linestyle='--', label='valid_tmp_mrr')

# Setting labels, title, and legend
plt.xlabel('Epoch')
plt.ylabel('MRR')
plt.title('MRR vs. Epoch')
plt.legend()
plt.grid(False)
plt.show()
```
